```python
from math import exp, sqrt
import matplotlib.pyplot as plt
from random import normalvariate as norm
```

```python
r = .0525
T = 1/252
sigma = .30
aapl = 187
prices = list()
```

```python
for day in range(63):
    prices.append(aapl)
    aapl *= exp((r - .5 * sigma ** 2)*T + sigma * sqrt(T) * norm(0,1))
```

```python
prices[:10]
```

```python
plt.plot(prices, label = "AAPL", ls = "-.")
plt.title("Possible Price Path for AAPL", fontsize= 14)
plt.ylabel("Price")
plt.xlabel("Days")
plt.grid(axis = "y", alpha = .5)
plt.legend(loc=2);
```
